```python
# Importing packages

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sbn
from sklearn.linear_model import LogisticRegression
```

```python
# reading train.csv
titanic_train = pd.read_csv("/kaggle/input/titanic/train.csv")
titanic_train = pd.DataFrame(titanic_train)
titanic_train.head(n=10)
```

```python
# train.csv  variable information
titanic_train.info()
```

```python
# train.csv statistics
titanic_train.describe()
```

```python
# checking null values
titanic_train.isnull().sum()
```

```python
# imputing null values of int variables with the mean
titanic_train['Age'].fillna(int(titanic_train['Age'].mean()), inplace=True)
```

```python
# creating dummies for ctegorical variables
titanic_train = pd.get_dummies(titanic_train, columns = ['Sex','Embarked'], dtype=float)
```

```python
# checking results
titanic_train.head(n=10)
```

```python
# reading test.csv
titanic_test = pd.read_csv('/kaggle/input/titanic/test.csv')
titanic_test = pd.DataFrame(titanic_test)
titanic_test.head(n=10)
```

```python
# test.csv  variable information
titanic_test.info()
```

```python
# test.csv statistics
titanic_test.describe()
```

```python
# checking null values
titanic_test.isnull().sum()
```

```python
# imputing null values of int variables with the mean
titanic_test['Age'].fillna(int(titanic_test['Age'].mean()), inplace=True)
titanic_test['Fare'].fillna(int(titanic_test['Fare'].mean()), inplace=True)
```

```python
# creating dummies for ctegorical variables
titanic_test = pd.get_dummies(titanic_test, columns = ['Sex','Embarked'], dtype=float)
```

```python
# checking results
titanic_test.head(n=10)
```

```python
# building the model using LogisticRegression() for the test dataset
model = LogisticRegression()
x_train = titanic_train.drop(['PassengerId', 'Cabin', 'Name', 'Ticket', 'Survived'], axis = 1)
y_train = titanic_train['Survived']
x_test = titanic_test.drop(['PassengerId', 'Cabin', 'Name', 'Ticket'], axis = 1)
model.fit(x_train, y_train)
y_pred = model.predict(x_test)
print(f"predicted response:\n{y_pred}")
```

```python
# calculating r-squared
r_sq = model.score(x_train, y_train)
print(f"R-Squared:{r_sq}")
```

```python
# creating target column from the predicted target obtained from the model
titanic_test['Survived'] = y_pred.tolist()
titanic_test.head(n=10)
```

```python
# preparing data for submission
titanic_submission = titanic_test[['PassengerId', 'Survived']]
titanic_submission.head(n=10)
```
